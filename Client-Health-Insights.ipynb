```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dados = pd.read_csv('seguro_de_vida.csv')

dados.head()
```

# Gráfico de Distribuição de Idade

```python
plt.figure(figsize=(8,5))
sns.histplot(dados['idade'], bins=20, kde=True, color='blue')
plt.title('Distribuição de Idade dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

# Boxplot de IMC por Status de Fumante

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='fumante', y='imc', data=dados, hue='fumante', palette='coolwarm', dodge=False)
plt.title('IMC por Status de Fumante')
plt.xlabel('Fumante')
plt.ylabel('IMC')
plt.legend([],[], frameon=False)  # Remove a legenda
plt.show()
```

# Gráfico de Contagem de Filhos

```python
plt.figure(figsize=(8,5))
sns.countplot(x='filhos', data=dados, hue='filhos', palette='viridis')
plt.title('Número de Filhos por Cliente')
plt.xlabel('Número de Filhos')
plt.ylabel('Contagem')
plt.legend([],[], frameon=False)  # Remove a legenda
plt.show()
```

# Gráfico de Distribuição das Despesas Médicas

```python
plt.figure(figsize=(8,5))
sns.histplot(dados['despesas'], bins=30, kde=True, color='green')
plt.title('Distribuição das Despesas Médicas')
plt.xlabel('Despesas')
plt.ylabel('Frequência')
plt.show()
```

# Gráfico de Dispersão: Idade vs. Despesas Médicas

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='idade', y='despesas', data=dados, hue='fumante', palette='coolwarm')
plt.title('Idade vs. Despesas Médicas')
plt.xlabel('Idade')
plt.ylabel('Despesas Médicas')
plt.show()
```

# Gráfico de Contagem de Clientes por Região

```python
plt.figure(figsize=(8,5))
sns.countplot(x='regiao', data=dados, hue='regiao', palette='Set2', legend=False)
plt.title('Distribuição de Clientes por Região')
plt.xlabel('Região')
plt.ylabel('Contagem')
plt.show()
```

# Gráfico Violin de Despesas Médicas por Fumante

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='fumante', y='despesas', data=dados, hue='fumante', palette='coolwarm', legend=False)
plt.title('Comparação de Despesas Médicas entre Fumantes e Não Fumantes')
plt.xlabel('Fumante')
plt.ylabel('Despesas Médicas')
plt.show()
```
